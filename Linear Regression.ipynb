```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\santosh nandam\Downloads\height_weight.csv")
data
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.describe()
```

```python
data.columns
```

```python
data.info()
```

```python
data.head(10)
```

```python
data.tail(10)
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data['Height'].isnull().sum()
```

```python
#split data
x=data.iloc[:,0:1].values
y=data.iloc[:,1].values
```

```python
x
```

```python
y
```

```python
#missing values detection in array
np.isnan(x).sum()
```

```python
np.isnan(y).sum()
```

```python
#fixing the missing values
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
x=imputer.fit_transform(x)
```

```python
np.isnan(x).sum()
```

```python
#overfit: training the data only with one condition where the model is not accurate.
#underfit: when a data is having mixed models and the training data is on a single condition, the testing data will also show the result as the trained data even if it is not in that condition.
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=.20, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
pred=LR.predict(x_test)
pred
```

```python
#visualize the output
plt.title('Height and Weight correlation on training data')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(x_train, y_train, color='lavender')
plt.plot(x_train,LR.predict(x_train),color='grey')
plt.show()
```

```python
#prediction of my weight
my_ht=[[185]]
my_wt_pred=LR.predict(my_ht)
```

```python
my_wt_pred
```

```python
#calculate R square
from sklearn.metrics import mean_squared_error
r2_score=LR.score(x,y)
r2_score
```
