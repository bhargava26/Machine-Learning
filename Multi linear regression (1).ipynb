```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\santosh nandam\Downloads\50_Startups_EDA.csv")
```

```python
dataset
```

```python
dataset.shape
```

```python
X=dataset.loc[:,["R&D Spend","Sales","Administration","Marketing Spend","State"]]
Y=dataset.loc[:,"Profit"]
```

```python
X
```

```python
Y
```

```python
df_1 = pd.get_dummies(X,['State'])
df_1
```

```python
df_1[['State_Chennai','State_Coimbatore','State_Trichy']]=df_1[['State_Chennai','State_Coimbatore','State_Trichy']].astype(int)
df_1
```

```python
df_1["R&D Spend"].fillna(df_1["R&D Spend"].mean())
df_1['Sales'].fillna(df_1['Sales'].mean())
df_1['Administration'].fillna(df_1['Administration'].mean())
df_1['Marketing Spend'].fillna(df_1['Marketing Spend'].mean())
df_1
```

```python
df_1.fillna(df_1.mean(),inplace=True)
df_1.isnull().sum()
```

```python
df_1=df_1.values
```

```python
df_1
```

```python
Y=Y.values
```

```python
from sklearn.model_selection import train_test_split
df_1_train, df_1_test, Y_train, Y_test=train_test_split(df_1,Y,test_size=.10, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(df_1_train,Y_train)
```

```python
Y_pred = LR.predict(df_1_test)
Y_pred
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(Y_test, Y_pred, color='blue', alpha=0.6)
plt.plot(Y_test, Y_test, color='red', linestyle='dotted') 
plt.xlabel("Actual Profit")
plt.ylabel("Predicted Profit")
plt.title("Actual vs. Predicted Profit (Multilinear Regression)")
plt.grid(True)
plt.show()
```

```python
my_input = np.array([[50000, 200000, 30000, 40000, 10000, 50000, 25000]])
profit=LR.predict(my_input)
profit
```

```python
from sklearn.metrics import mean_squared_error
r2_score=LR.score(df_1,Y)
r2_score
```
